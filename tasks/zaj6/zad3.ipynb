```python
import pandas as pd
%matplotlib inline
from matplotlib import pylab
```

```python
data = pd.read_csv('data-errors.csv', header=2)
```

```python
data['Rok'].fillna(method='ffill', inplace=True)
data[['Rok', 'Tydzień']] = data[['Rok', 'Tydzień']].astype(int)

data.dtypes
```

```python
from datetime import date, timedelta
def week_start_date(year, week, dow=0):    
    """
    :param int year: Numer roku
    :param int week: Numer tygodnia w roku 
    :param int dow: Funkcja ta zwraca datę, zwróci datę dnia o indeksie dow, zero oznacza poniedziałek, jeden wtorek itp. 
    """
    # http://stackoverflow.com/a/1287862/7918
    d = date(year, 1, 1)    
    delta_days = d.isoweekday() - 1
    delta_weeks = week
    if year == d.isocalendar()[0]:
        delta_weeks -= 1
    delta = timedelta(days=-delta_days+dow, weeks=delta_weeks)
    return d + delta
```

```python
data.drop('Unnamed: 0', 1, inplace=True)
```

```python
dates = [week_start_date(int(y), int(w)) for y, w in zip(data['Rok'], data['Tydzień'])]

data_i = data.set_index(pd.DatetimeIndex(dates), verify_integrity=True)
```
